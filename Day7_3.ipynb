```python
# -----------------------------------------------
# Mini Chatbot using Python
# Rule-Based Chatbot using Tokenization
# -----------------------------------------------

import nltk
from nltk.tokenize import word_tokenize
import random

# Download tokenizer data (Run only once)
nltk.download('punkt')
nltk.download('punkt_tab')

print("=" * 60)
print("           MINI CHATBOT USING PYTHON")
print("=" * 60)

print("Type 'exit' to stop the chatbot.\n")

# Knowledge Base
responses = {

    "hello": [
        "Hello! How can I help you?",
        "Hi! Nice to meet you.",
        "Greetings!"
    ],

    "hi": [
        "Hello!",
        "Hi there!",
        "Welcome!"
    ],

    "python": [
        "Python is a high-level programming language.",
        "Python is widely used in Artificial Intelligence and Machine Learning."
    ],

    "ai": [
        "Artificial Intelligence enables machines to learn and make decisions.",
        "AI is one of today's fastest-growing technologies."
    ],

    "artificial": [
        "Artificial Intelligence enables machines to learn and make decisions."
    ],

    "machine": [
        "Machine Learning is a subset of Artificial Intelligence.",
        "Machine Learning allows computers to learn from data."
    ],

    "learning": [
        "Machine Learning is a branch of Artificial Intelligence."
    ],

    "weather": [
        "I cannot access live weather information.",
        "Please check an online weather service."
    ],

    "name": [
        "I am a Mini Chatbot developed in Python."
    ],

    "bye": [
        "Goodbye!",
        "Have a great day!",
        "See you again!"
    ]
}

# Chat Loop
while True:

    user = input("\nYou : ").lower()

    if user == "exit":
        print("Bot : Goodbye!")
        break

    # Tokenization
    tokens = word_tokenize(user)

    found = False

    # Keyword Matching
    for word in tokens:

        if word in responses:
            print("Bot :", random.choice(responses[word]))
            found = True
            break

    # Default Response
    if not found:
        print("Bot : Sorry, I don't understand. Please ask another question.")
```
