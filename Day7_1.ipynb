```python
import nltk
from nltk.tokenize import sent_tokenize

# Download tokenizer data (run only once)
nltk.download('punkt')
nltk.download('punkt_tab')

paragraph = """
Artificial Intelligence is growing rapidly.
Machine Learning is a subset of AI.
Deep Learning uses neural networks.
Students enjoy learning NLP.
"""

sentences = sent_tokenize(paragraph)

print("Paragraph:\n")
print(paragraph)

print("\nSentences:\n")

for i, sentence in enumerate(sentences, start=1):
    print(f"Sentence {i}: {sentence}")
```

```python
import spacy

# Load the English language model
nlp = spacy.load("en_core_web_sm")

# Input text
text = "OpenAI develops powerful AI models like GPT."

# Process the text
doc = nlp(text)

print("Tokens:\n")

# Print each token and its Part-of-Speech (POS) tag
for token in doc:
    print(f"{token.text:15} -> {token.pos_}")
```

```python
from transformers import AutoTokenizer

# Load the BERT tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Input text
text = "Artificial Intelligence is amazing."

# Tokenize the text into subword tokens
tokens = tokenizer.tokenize(text)

# Convert tokens to token IDs
token_ids = tokenizer.convert_tokens_to_ids(tokens)

# Display the results
print("Original Text:")
print(text)

print("\nSubword Tokens:")
print(tokens)

print("\nToken IDs:")
print(token_ids)

print("\nDecoded Text:")
print(tokenizer.decode(token_ids))
```
