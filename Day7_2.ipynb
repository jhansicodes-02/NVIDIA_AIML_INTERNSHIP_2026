```python
def chatbot():
    print("ChatBot: Hello! Type 'bye' to exit.")

    while True:
        user = input("You: ").lower()

        if user == "hello":
            print("ChatBot: Hi! How are you?")
        elif user == "how are you":
            print("ChatBot: I'm doing great!")
        elif user == "what is your name":
            print("ChatBot: I'm a simple Python chatbot.")
        elif user == "bye":
            print("ChatBot: Goodbye!")
            break
        else:
            print("ChatBot: Sorry, I don't understand.")

chatbot()
```

```python
import nltk
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('punkt_tab')

responses = {
    "hello": "Hi!",
    "ai": "Artificial Intelligence is the simulation of human intelligence.",
    "python": "Python is a popular programming language."
}

while True:
    user = input("You: ")

    if user.lower() == "bye":
        print("Bot: Goodbye!")
        break

    tokens = word_tokenize(user.lower())

    found = False

    for token in tokens:
        if token in responses:
            print("Bot:", responses[token])
            found = True
            break

    if not found:
        print("Bot: I don't know about that.")
```

```python
from transformers import pipeline

chatbot = pipeline(
    "text2text-generation",
    model="google/flan-t5-base"
)

while True:
    user = input("You: ")

    if user.lower() == "bye":
        break

    response = chatbot(user, max_new_tokens=50)

    print("Bot:", response[0]["generated_text"])
```

```python
from openai import OpenAI

client = OpenAI(api_key="YOUR_API_KEY")

while True:
    user = input("You: ")

    if user.lower() == "bye":
        break

    response = client.responses.create(
        model="gpt-5.5",
        input=user
    )

    print("Bot:", response.output_text)
```
